```python
from pandas_data_reader_service_core.modules.finam.search_service import Search, Market
from pandas_data_reader_service_core.modules.finam.stock_info import FinamStockInfo
from pandas_data_reader_service_core.service import PandasDataReaderService as pdrs, StockInfo, TimeFrame

from datetime import date

import plotly.graph_objects as go

import pandas as pd
from src.ElliotWaves.waves.impulse_five_search import search
```

```python
from __future__ import annotations
import logging
import warnings
logging.basicConfig(format='\033[93m%(asctime)s %(levelname)s %(name)s:\033[0m %(message)s', level=logging.INFO, datefmt='%I:%M:%S')
warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)
warnings.simplefilter(action="ignore", category=pd.core.common.SettingWithCopyWarning)
```

```python
stock = "EURUSD"
market = Market.CURRENCIES_WORLD

date_from = date(2015,1,1)
date_till = date(2021,1,1)
tf = TimeFrame.DAILY
```

```python
stock_list = Search.by_code(stock, market)
stock_list
```

```python
stock_info = StockInfo.init_for_Finam(FinamStockInfo(stock_list.index[0]))
print(stock_info)
```

```python
data_df = pdrs().get(stock=stock_info,  date_from=date_from,date_to=date_till,time_frame=tf,short_col_name=True)
data_df
```

```python
live_waves,dead_waves = search(data_df, full_scan=True)
```

```python
from example.ElliotWales.misc.analitic import analitic_df
```

```python
waves_df = analitic_df(live_waves).append(analitic_df(dead_waves),ignore_index=True).drop("SubWaves",axis=1)
waves_df
```

```python
import os
file_name = f'Analitic_{stock}_{date_from.strftime("%Y%m%d")}_{date_till.strftime("%Y%m%d")}_{tf.name}.csv'
output_dir = "files"
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
file_path = os.path.join(output_dir,file_name)
if os.path.exists(file_path):
    os.remove(file_path)
waves_df.to_csv(file_path,index_label="index")
```
