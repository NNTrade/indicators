# Analitic

```python
from src.ElliotWaves.waves.impulse_five_search import condens_sub_waves
from src.RelativePercent.Factory import get_percent
from example.ElliotWales.misc.analitic import analitic_df
```

```python
def analitic_tmp_df(waves:List[wave])->pd.DataFrame:
    dir_col_name = "Direction"
    sub_waves_arr_col_name = "SubWaves"
    ret_waves_df =pd.DataFrame(pd.Series(condens_sub_waves(waves),name=sub_waves_arr_col_name))
    ret_waves_df["SubWaveCount"] = ret_waves_df[sub_waves_arr_col_name].map(lambda el: len(el))
    ret_waves_df[dir_col_name] = ret_waves_df[sub_waves_arr_col_name].map(lambda el: el[0].direction)
    ret_waves_df["Full"] = ret_waves_df["SubWaveCount"] == 5
    for idx in range(5):
        col_name = f"Wave{idx+1} Profit%"
        waves_sr:pd.Series = ret_waves_df[ret_waves_df["SubWaveCount"] >= (idx+1)][sub_waves_arr_col_name]
        wave_sr:pd.Series = waves_sr.map(lambda el: el[idx])
        
        ret_waves_df[col_name] = get_percent(sr1=wave_sr.map(lambda el: el.end.price),sr2=wave_sr.map(lambda el: el.start.price)) * wave_sr.map(lambda el: (-1 if el.direction == direction.Short else 1 ))
    return ret_waves_df
```

```python
tmp_df = analitic_tmp_df(dead_waves)
tmp_df
```

```python
dead_waves_df = analitic_df(dead_waves)
dead_waves_df
```

```python
live_waves_df = analitic_df(live_waves)
live_waves_df
```

```python
waves_df = live_waves_df.append(dead_waves_df,ignore_index=True).drop("SubWaves",axis=1)
waves_df
```

```python
import os
file_name = f'Analitic_{stock}_{date_from.strftime("%Y%m%d")}_{date_till.strftime("%Y%m%d")}_{tf.name}.csv'
output_dir = "files"
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
file_path = os.path.join(output_dir,file_name)
if os.path.exists(file_path):
    os.remove(file_path)
waves_df.to_csv(file_path,index_label="index")
```
